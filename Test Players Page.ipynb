```python
import pandas as pd
```

```python
players = pd.read_csv('Data/cfbstats-com-2009/player.csv')

players.head()
```

```python
players.drop(columns=['Uniform Number', 'Class', 'Position', 'Height', 'Weight', 
                      'Home Town', 'Home State', 'Home Country', 'Last School'], 
             axis=1, 
             inplace=True)

players.head()
```

```python
player_stats = pd.read_csv('Data/cfbstats-com-2009/player-game-statistics.csv')

player_stats.head()
```

```python
player_stats.drop(columns='Game Code', axis=1, inplace=True)
```

```python
player_stats = player_stats.groupby('Player Code').sum()

player_stats.head()
```

```python
team = pd.read_csv('Data/cfbstats-com-2009/team.csv')

team.head()
```

```python
new_players = pd.merge(players, player_stats, on='Player Code')

new_players.head()
```

```python
new_players = pd.merge(new_players, team, on='Team Code')

new_players.head()
```

```python
new_players.drop(columns=['Player Code', 'Team Code', 'Conference Code'], axis=1, inplace=True)

new_players.head()
```

```python
combine = pd.read_csv('Data/NFL.csv')

combine.head()
```

```python
combine['Player'] = combine['Player'].str.split('\\', expand=True)

combine.head()
```

```python
combine['Last_Name'] = combine['Player'].str.split(' ', expand=True)[1]

combine.head()
```

```python
new_players
```

```python
final = pd.merge(new_players, combine, left_on=['Last Name', 'Name'], right_on=['Last_Name', 'School'])

final.head()
```
